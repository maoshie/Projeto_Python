```python
#add
def add(x,y):
    return (x+y)

#subtract
def subtract(x,y):
    return (x-y)

#multiply
def multiply(x,y):
    return (x*y)

#divide
def divide(x,y):
    return (x/y)


#2 - Mostrar na tela o numero da opção;
print("\nSelecione o número da operação desejada: ")
print("1 - Soma")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")


#3 - Mostrar(Digite sua opção);
escolha = input("\nDigite a opção desejada e aperte enter: ")


#4 - Declarando a variavel pra receber os numeros digitados;
num1 = int(input("\nDigite o primeiro numero: "))
num2 = int(input("\nDigite o segundo numero: "))



#2 - Mostrar na tela o numero da opção;
print("\nSelecione o número da operação desejada: ")
print("1 - Soma")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")


#5 - Montar o calculo:
#5.1 add
if escolha == "1":
    print("\n")
    print(num1, "+", num2, "=", add(num1,num2))
    print("\n")


#5.2 subtract
elif escolha == "2":
    print("\n")
    print(num1, "-", num2, "=", subtract(num1,num2))
    print("\n")


#5.3 Multiply
elif escolha == "3":
    print("\n")
    print(num1,"*",num2,"=", multiply(num1,num2))
    

#5.4 Divide 
elif escolha == "4":
    print("\n")
    print(num1,"/",num2,"=",divide(num1,num2))
    
    
#5.5 erro
else:
    print("\n")
    print("opção invalida")
```
